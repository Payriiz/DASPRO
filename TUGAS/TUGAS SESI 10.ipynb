```python
#kumpulan data
daftar_mahasiswa = [] #list
mahasiswa_dict = {} #dictionary

def tambah():
    nama = input("MASUKKAN NAMA: ")
    nim = int(input("MASUKKAN NIM: "))
    jurusan = input("MASUKKAN JURUSAN: ")
    ipk = float(input("MASUKKAN IPK: "))

    mahasiswa = (nama,nim,jurusan,ipk)
    daftar_mahasiswa.append(mahasiswa)
    mahasiswa_dict[nim] = mahasiswa

    print("DATA BERHASIL DITAMBAHKAN!")

def tampil():
    if not daftar_mahasiswa:
        print("TIDAK ADA DATA MAHASISWA!")
    else:
        for daftar in daftar_mahasiswa:
            print(f"NIM:{daftar[1]} \t NAMA:{daftar[0]} \t JURUSAN:{daftar[2]} \t IPK:{daftar[3]}")

def cari():
    pencarian = int(input("MASUKKAN NIM YANG DICARI: "))
    if pencarian in mahasiswa_dict:
        hasil = mahasiswa_dict[nim]
        print(f"""
        DATA DITEMUKAN!:
        NIM:{hasil[1]} \t NAMA:{hasil[0]} \t JURUSAN:{hasil[2]} \t IPK:{hasil[3]}
        """)
    else:
        print("DATA TIDAK DITEMUKAN")


def update():
     nim = int(input("MASUKKAN NIM YANG AKAN DIUBAH IPKNYA: "))
     if nim in mahasiswa_dict:
         print(f"NAMA:{mahasiswa_dict[nim][0]} \t JURUSAN:{mahasiswa_dict[nim][2]} \t IPK{mahasiswa_dict[nim][3]}")
         ipk_baru = float(input("MASUKKAN IPK BARU: "))
         mahasiswa_update = (nama,nim,jurusan,ipk_baru)
         daftar_mahasiswa.remove(mahasiswa_dict[nim])
         daftar_mahasiswa.append(mahasiswa_update)
         mahasiswa_dict[nim] = mahasiswa_update
         print("IPK BERHASIL DIPERBAHARUI!")
     else:
         print("TIDAK ADA DATA MAHASISWA")

def hapus():
    nim = int(input("MASUKKAN NIM YANG AKAN DIHAPUS: "))
    if nim in mahasiswa_dict:
        daftar_mahasiswa.remove(mahasiswa_dict[nim])
        del mahasiswa_dict[nim]
        print("DATA BERHASIL DIHAPUS!")
    else:
        print("DATA TIDAK DITEMUKAN!")
```

```python
while True:
    print("""
    ---MENU---
    1. TAMBAH MAHASISWA
    2. TAMPILKAN SELURUH MAHASISWA
    3. CARI MAHASISWA (dengan NIM)
    4. UPDATE IPK MAHASISWA
    5. HAPUS MAHASISWA
    6. KELUAR
    """)

    pilihan = int(input("MASUKKAN PILIHAN (1-6): "))

    if pilihan==1:
        tambah()
    elif pilihan==2:
        tampil()
    elif pilihan==3:
        cari()
    elif pilihan==4:
        update()
    elif pilihan==5:
        hapus()
    elif pilihan==6:
        print("ANDA KELUAR DARI PROGRAM")
        break
```
